```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
model=tf.keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
# we have created a neural network with one layer and one neuron in that layer
```

```python
model.compile(optimizer='sgd',loss='mean_squared_error')
# we have specified how the optimizer should work and how the loss function should work i.e. which method should be used to find next guess for optimizer
# and the differnce between actual result from the training data with the guess taken to find the answer i.e. the equation of y=mx+c where m and c values
# are unkown for the below data
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

```python
model.fit(xs,ys,epochs=500)
```

```python
print(model.predict(np.array([10.0])))
```
